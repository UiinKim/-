```python
from google.colab import files
uploaded=files.upload()
```

```python
import pandas as pd

dataframes = [pd.read_csv(file_path, encoding='utf-8') for file_path in uploaded]
df = pd.concat(dataframes, ignore_index=True)
```

```python
import numpy as np

df['제품군'] = '로션'

df['review'] = df['review'].astype(str)

# 남성 컬럼 추가
def check_male(text):
    keywords = ['남자', '아빠', '남편이', '남편', '남성', '오빠', '남자친구', '남동생']
    if any(keyword in text for keyword in keywords):
        return 1
    return np.nan

df['남성'] = df['review'].apply(check_male)

# 수분 컬럼 추가
def check_moisture(text):
    positive_keywords = ['수분', '촉촉', '보습', '안개', '미스트', '뿌려', '뿌리']
    negative_keyword = '건조'
    if any(keyword in text for keyword in positive_keywords):
        return 1
    elif negative_keyword in text:
        return 0
    return np.nan

df['수분'] = df['review'].apply(check_moisture)

# 향 컬럼 추가
def check_fragrance(text):
    keywords = ['향이', '향도', '향', '향은']
    if any(keyword in text for keyword in keywords):
        return 1
    return np.nan

df['향'] = df['review'].apply(check_fragrance)

# 자극 컬럼 추가
def check_irritation(text):
    irritant_keywords = ['자극', '여드름', '예민한', '트러블', '민감']
    soothing_keywords = ['진정', '순하고', '순해서']
    if any(keyword in text for keyword in soothing_keywords):
        return 1
    elif any(keyword in text for keyword in irritant_keywords):
        return 0
    return np.nan

df['자극'] = df['review'].apply(check_irritation)

# 지속성 컬럼 추가
def check_duration(text):
    if '오래' in text:
        return 1
    elif '금방' in text:
        return 0
    return np.nan

df['지속성'] = df['review'].apply(check_duration)

# 계절 컬럼 추가
def check_season(text):
    if '여름' in text:
        return 1
    elif '겨울' in text:
        return 0
    return np.nan

df['계절'] = df['review'].apply(check_season)

# 유분 컬럼 추가
def check_oil(text):
    if '유분' in text:
        return 1
    return np.nan

df['유분'] = df['review'].apply(check_oil)

# 피부타입 컬럼 추가
def check_skin_type(text):
    if '지성' in text:
        return 1
    elif '건성' in text:
        return 2
    elif '복합성' in text:
        return 0
    return np.nan

df['피부타입'] = df['review'].apply(check_skin_type)
```

```python
# 결과 저장
df.to_csv('스킨케어_로션_labeling.csv', index=False, encoding='cp949')
```
