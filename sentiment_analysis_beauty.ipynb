```python
from google.colab import files
uploaded=files.upload()
```

```python
  # import library
  import torch
  from transformers import AutoTokenizer, AutoModelForSequenceClassification, TextClassificationPipeline

  # load model
  tokenizer = AutoTokenizer.from_pretrained("jaehyeong/koelectra-base-v3-generalized-sentiment-analysis")
  model = AutoModelForSequenceClassification.from_pretrained("jaehyeong/koelectra-base-v3-generalized-sentiment-analysis")
  sentiment_classifier = TextClassificationPipeline(tokenizer=tokenizer, model=model)
```

```python
def preprocess_review(review, max_length=510):
    inputs = tokenizer.encode_plus(review, add_special_tokens=True, max_length=max_length, truncation=True, return_tensors="pt")
    return tokenizer.decode(inputs['input_ids'][0])
```

```python
import pandas as pd

files = [pd.read_csv(file_path, encoding='cp949') for file_path in uploaded]
i=1
for file in files:
  print('%d start'%(i))
  labels=[]
  scores=[]
  for review in file['review']:
    print(len(review))
    review = preprocess_review(review)
    pred = sentiment_classifier(review)
    labels.append(pred[0]['label'])
    scores.append(pred[0]['score'])
  file['sentiment_label']=labels
  file['sentiment_score']=scores
  file.to_csv('sentiment_labeling_%d.csv'%(i), index=False)
  i+=1
```
